```python
def check_loan(row):
    if row['balance'] < 0 or row['loan'] == 'yes' and row['job'] == 'unemployed':
        return "No Credit"
    else:
        return "Gives Credit"

df['status'] = df.apply(check_loan, axis=1)
```

```python
#in this code we create bar which shows that which type of jobs are potensial for credit

import matplotlib.pyplot as plt

df = pd.read_csv('bank.csv', sep=';')

potensial = df[
    (df['balance'] >500) &
    (df['loan'] == 'no') &
    (df['default'] == 'no')
]
print(df)

print("Job types:")
print(potensial['job'].value_counts())

potensial['job'].value_counts().plot(kind='bar', color='blue')
plt.title('Potensial Customers —Job types')

plt.show()
```

```python
#In this code we write code for which months are best for suggest to customers credit due to their "yes" answers.

month_result = df.groupby('month')['y'].apply(
    lambda x: (x == 'yes').mean() * 100
).round()

month_desc_result=month_result.sort_values(ascending=False).head(3)
print(month_desc_result)
```

```python
#We show statistic using pie
month_desc_result.plot(kind='pie')
plt.title('Success persentage (%)')
plt.show()
```

```python
#there are students,which have a lot balance .So they can be good customers for the bank in the future
rich_students = df[(df['job'] == 'student') & (df['balance'] > 3000)]
display(rich_students)
```

```python
#There are good customers who don't have any lateness for pay credit
good_customers=df[(df['default'] == 'no') & (df['loan']=='yes')]
display(good_customers)
```

```python
#There are risky users,who have loan more than one and balance is less than 1000
risky_users = df[(df['balance'] <1000) & (df['housing'] == 'yes') & (df['loan'] == 'yes')]
display(risky_users)
```

```python
too_many_calls = df[(df['campaign'] > 10) & (df['y'] == 'no')]
display(too_many_calls)
```

```python
total_bank_balance = df['balance'].sum()
average_balance = df['balance'].mean()
total_customers = len(df)

customers_with_loans = df[df['loan'] == 'yes'].shape[0]
loan_percentage = (customers_with_loans / total_customers) * 100


# High Value: Balance > 5000
# Mid Value: 1000 < Balance <= 5000
# Low Value: Balance <= 1000
high_value_customers = df[df['balance'] > 5000].shape[0]
mid_value_customers = df[(df['balance'] > 1000) & (df['balance'] <= 5000)].shape[0]
low_value_customers = df[df['balance'] <= 1000].shape[0]


successful_calls = df[df['y'] == 'yes'].shape[0]
success_rate = (successful_calls / total_customers) * 100


print("==================================================")
print("          BANK FINANCIAL SUMMARY REPORT           ")
print("==================================================")
print(f"Total Capital in Bank      : ${total_bank_balance:,}")
print(f"Average Customer Balance   : ${average_balance:.2f}")
print(f"Total Number of Customers  : {total_customers}")
print("--------------------------------------------------")
print(f"Customers with Personal Loan: {customers_with_loans} ({loan_percentage:.1f}%)")
print(f"Campaign Success Rate      : {success_rate:.1f}%")
print("--------------------------------------------------")
print("CUSTOMER SEGMENTATION:")
print(f" - High Value (>$5000)     : {high_value_customers} users")
print(f" - Mid Value ($1000-$5000) : {mid_value_customers} users")
print(f" - Low Value (<$1000)      : {low_value_customers} users")
print("==================================================")

print("\nTOP 5 BEST JOBS (By Average Balance):")
top_jobs = df.groupby('job')['balance'].mean().sort_values(ascending=False).head(5)
print(top_jobs)
```
